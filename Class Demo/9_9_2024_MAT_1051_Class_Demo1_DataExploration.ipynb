```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#defining our dataframe
videos = pd.read_csv("https://raw.githubusercontent.com/fedhere/FDSFE_FBianco/main/data/USyoutubes.csv")
```

```python
videos.head()
```

```python
#subset of the data
videos_restrict = videos[["views", "likes", "dislikes", "comment_count"]]
```

```python
videos_restrict
```

```python
videos_restrict.corr()
```

```python
#clim is column limit
plt.imshow(videos_restrict.corr(), clim=(-1,1), cmap="RdBu")
#gives the range of value
plt.colorbar()
#the semi-column is in place because the notebook returns the last value of the previous line. The semi-column surprisees the output
plt.xticks([0,1,2,3], videos_restrict.columns, rotation= 45);
plt.xticks([0,1,2,3], videos_restrict.columns, rotation= 45);
```

```python
#shows the first 5 rows
videos_restrict.head()
```

```python
#the last five rows
videos_restrict.tail()
```

```python
#getting the max number of likes
videos_restrict["likes"].max()
```

```python
videos_restrict["likes"].min()
```

```python
max_likes = videos_restrict["likes"].max()
videos[videos["likes"] == max_likes]
```

```python
#the video with the most dislikes
#storing the max value of dislikes column in a variable
max_dislikes = videos_restrict["dislikes"].max()
#using the variable to idex the dataset for the row which has the most number of dislikes
videos[videos["dislikes"] == max_dislikes]
```

```python
#the video with the least number of likes
min_likes = videos_restrict["likes"].min()
videos[videos["likes"] == min_likes]
```
